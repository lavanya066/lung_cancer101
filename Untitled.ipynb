```python
dir=pathlib.Path(dir)
dir
```

```python
list(dir.glob('Bengin cases/*'))
```

```python
lung_dict={
    'bengin':list(dir.glob('Bengin cases/*')),
    'malign':list(dir.glob('Malignant cases/*')),
    'normal':list(dir.glob('Normal cases/*'))
}
```

```python
lungs_label={
    'bengin':0,
    'malign':1,
    'normal':2
}
```

```python
img=cv2.imread(str(lung_dict['bengin'][0]))
plt.imshow(img)
```

```python
img.show()
```

```python
x=[]
y=[]
for name,img in lung_dict.items():
    for image in img:
        im=cv2.imread(str(image))
        im=cv2.resize(im,(120,120))
        x.append(im)
        y.append(name)
```

```python
len(x)
```

```python
print(x[500],y[500])
```

```python
plt.imshow(x[500])
plt.xlabel(y[500])
```

```python
for i in range(len(y)):
    y[i]=lungs_label[y[i]]
```

```python
x=np.array(x)
y=np.array(y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
x_train=x_train/255

x_test=x_test/255
```

```python
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras import datasets,layers,models
```

```python
model=models.Sequential([
    layers.Conv2D(32,3,padding="same",activation='relu'),
    layers.MaxPooling2D(),
    layers.Conv2D(32,3,padding="same",activation='relu'),
    layers.MaxPooling2D(),
    layers.Conv2D(64,3,padding="same",activation='relu'),
    layers.MaxPooling2D(),
    layers.Flatten(),
    layers.Dense(128,activation='relu'),
    layers.Dense(3,activation='softmax')
    
])
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy']
             )
model.fit(x_train,y_train,epochs=10)
```

```python
y_pred=model.predict(x_test)
y_pred[0]
```

```python
r=np.argmax(y_pred[0])
for key,val in lungs_label.items():
    if(val==r):
        print(key)
        break
```
